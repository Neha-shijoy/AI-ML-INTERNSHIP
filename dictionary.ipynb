1. Create and Access
 Create a dictionary student with keys name, age, and grade.
 • Print grade using both [] and .get().
 • Try .get() for a non-existing key and print “Key not found”.

```python
student = {"name": "John", "age": 20, "grade": "A"}

print(student["grade"])
print(student.get("grade"))

print(student.get("city", "Key not found"))
```

3. pop()
 fruits = {"apple": 100, "banana": 60, "orange": 80}
 • Remove “banana” using .pop() and print the removed value.

```python
fruits = {"apple": 100, "banana": 60, "orange": 80}

removed = fruits.pop("banana")
print(removed)
print(fruits)
```

7. keys()
 country = {"India": "Delhi", "Japan": "Tokyo", "France": "Paris"} 
 • Print all keys using .keys()
 • Iterate and print keys one by one.

```python
country = {"India": "Delhi", "Japan": "Tokyo", "France": "Paris"}

print(country.keys())

for k in country:
    print(k)
```

8. values()
 • Print all values using .values() 
 • Store all values into a list.

```python
country = {"India": "Delhi", "Japan": "Tokyo", "France": "Paris"}

print(country.values())

values_list = list(country.values())
print(values_list)
```

9. items()
 student = {"name": "Asha", "age": 20, "grade": "A"} • Print all key-value pairs using .items() 
 • Iterate and print “Key: name, Value: Asha” format.

```python
student = {"name": "Asha", "age": 20, "grade": "A"}

print(student.items())

for k, v in student.items():
    print(f"Key: {k}, Value: {v}")
```

10. copy() 
 original = {"x": 10, "y": 20}
  • Make a copy using .copy() 
  • Modify copy and show original remains same.

```python
original = {"x": 10, "y": 20}

copy_dict = original.copy()
copy_dict["x"] = 99

print(original)
print(copy_dict)
```

14. Nested Dictionary
 students = {"Ravi": {"math": 80, "science": 85}, "Meena": {"math": 90, "science": 95}}
  • Print Ravi’s science marks.
  • Add new subject english: 88 for Meena.

```python
students = {
    "Ravi": {"math": 80, "science": 85},
    "Meena": {"math": 90, "science": 95}
}

print(students["Ravi"]["science"])

students["Meena"]["english"] = 88
print(students)
```

15. Dictionary Comprehension 
• Create dict of squares (1–10). 
• Convert lowercase letters to uppercase using comprehension.

```python
squares = {x: x*x for x in range(1, 11)}
print(squares)

letters = {chr(i): chr(i).upper() for i in range(97, 123)}
print(letters)
```

16. Combining Multiple Methods
 inventory = {} 
 • Add three items with quantities. 
 • Update one quantity.
 • Remove one item using .pop() 
 • Clear dictionary.

```python
inventory = {}

inventory["pen"] = 10
inventory["book"] = 5
inventory["pencil"] = 20

inventory["book"] = 8

inventory.pop("pen")

inventory.clear()

print(inventory)
```
