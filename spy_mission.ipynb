Spy Mission Management System (Python Project)


```python
import random

agents = []

# 🔹 1. Add Agent
def add_agent(name):
    agents.append({"name": name, "missions": []})
    print("Agent added successfully!")

# 🔹 2. Assign Mission Scores (*args used here)
def assign_mission(name, *scores):
    for agent in agents:
        if agent["name"].lower() == name.lower():
            agent["missions"].extend(scores)
            print("Mission scores assigned!")
            return
    print("Agent not found!")

# 🔹 3. View All Agents (**kwargs used here)
def view_agents(**kwargs):
    if len(agents) == 0:
        print("No agents available.")
        return

    for agent in agents:
        print("\nName:", agent["name"])
        print("Missions:", agent["missions"])

# 🔹 4. Calculate Performance
def performance(missions):
    if len(missions) == 0:
        return 0
    return sum(missions) / len(missions)

# 🔹 5. Rank System
def get_rank(avg):
    if avg >= 90:
        return "Elite Agent"
    elif avg >= 75:
        return "Senior Agent"
    elif avg >= 50:
        return "Junior Agent"
    else:
        return "Failed Agent"

# 🔹 6. Show Top Agent
def top_agent():
    if len(agents) == 0:
        print("No agents found.")
        return

    top = None
    top_score = 0

    for agent in agents:
        avg = performance(agent["missions"])
        if avg > top_score:
            top_score = avg
            top = agent

    print("\nTop Agent:", top["name"])
    print("Average Score:", top_score)
    print("Rank:", get_rank(top_score))

# 🔹 7. Generate Report
def report():
    print("\n--- Mission Report ---")

    for agent in agents:
        avg = performance(agent["missions"])
        rank = get_rank(avg)

        print("\nName:", agent["name"])
        print("Average:", avg)
        print("Rank:", rank)

# 🔹 BONUS: Add Random Mission Score
def bonus_mission(name):
    bonus = random.randint(5, 20)

    for agent in agents:
        if agent["name"].lower() == name.lower():
            agent["missions"].append(bonus)
            print("Bonus mission added:", bonus)
            return

    print("Agent not found!")

# 🔹 MENU SYSTEM
while True:
    print("\n===== SPY MISSION SYSTEM =====")
    print("1. Add Agent")
    print("2. Assign Mission")
    print("3. View Agents")
    print("4. Show Top Agent")
    print("5. Generate Report")
    print("6. Bonus Mission")
    print("7. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter agent name: ")
        add_agent(name)

    elif choice == "2":
        name = input("Enter agent name: ")
        scores = list(map(int, input("Enter scores separated by space: ").split()))
        assign_mission(name, *scores)

    elif choice == "3":
        view_agents()

    elif choice == "4":
        top_agent()

    elif choice == "5":
        report()

    elif choice == "6":
        name = input("Enter agent name: ")
        bonus_mission(name)

    elif choice == "7":
        print("Exiting system...")
        break

    else:
        print("Invalid choice!")
```
