Adding Elements (Methods Only) Create an empty list and add 5 numbers using append().

```python
my_list = []
my_list.append(1)
my_list.append(2)
my_list.append(3)
my_list.append(4)
my_list.append(5)
print(my_list)
```

Insert a value at index 2 using insert().

```python
my_list.insert(2, 10)
print(my_list)
```

Add multiple elements from another list using extend().

```python
other_list = [6, 7, 8]
my_list.extend(other_list)
```

Add a list inside another list using append() and observe the difference from extend().

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list1.extend(list2)
print(list1)
```
