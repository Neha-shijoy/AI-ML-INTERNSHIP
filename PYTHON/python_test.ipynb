6. Given: nums = [10, 20, 30, 20, 40] 
 • Count how many times 20 appears
 • Remove one occurrence of 20 
 • Add 50 to the list

```python
nums = [10, 20, 30, 20, 40]

print(nums.count(20))   # count occurrences

nums.remove(20)         # remove one occurrence
nums.append(50)         # add 50

print(nums)
```

10. Given: nums = [1,2,2,3,3,3,4]
• Count frequency using count() 
• Store result in dictionary

```python
nums = [1,2,2,3,3,3,4]

freq = {}

for i in nums:
    freq[i] = nums.count(i)

print(freq)
```

11. Given: list1 = [1,2,3] list2 = [3,4,5]
 • Combine lists 
 • Remove duplicates 
 • Sort final list

```python
list1 = [1,2,3]
list2 = [3,4,5]

combined = list1 + list2

unique = list(set(combined))

unique.sort()

print(unique)
```
