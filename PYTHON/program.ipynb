Write a program to check whether a number is positive, negative, or zero.

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

Create a program to check if a person is eligible to vote (age ≥ 18).

```python
age = int(input("Enter age: "))

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible")
```

Write a program to check whether a number is even or odd.

```python
num = int(input("Enter number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

Take a character input and check whether it is a vowel or consonant.

```python
ch = input("Enter a character: ").lower()

if ch in "aeiou":
    print("Vowel")
else:
    print("Consonant")
```

Write a program to find the largest of two numbers.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Largest:", a)
else:
    print("Largest:", b)
```

Write a program to find the largest of three numbers.

```python
a = int(input())
b = int(input())
c = int(input())

if a >= b and a >= c:
    print("Largest:", a)
elif b >= a and b >= c:
    print("Largest:", b)
else:
    print("Largest:", c)
```

Create a program to calculate grade based on marks:

90+ → A

75–89 → B

50–74 → C

Below 50 → Fail

```python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("A")
elif marks >= 75:
    print("B")
elif marks >= 50:
    print("C")
else:
    print("Fail")
```

Create a simple login system:

If username = "admin" and password = "1234" → Login success

Else → Invalid credentials

```python
username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin" and password == "1234":
    print("Login success")
else:
    print("Invalid credentials")
```

Write a program to check if a number is divisible by both 3 and 5.

```python
num = int(input("Enter number: "))

if num % 3 == 0 and num % 5 == 0:
    print("Divisible by both 3 and 5")
else:
    print("Not divisible")
```


Print the sum of first 10 natural numbers.

```python
sum = 0

for i in range(1, 11):
    sum += i

print("Sum =", sum)
```

Print numbers from 10 to 1 (reverse order).

```python
for i in range(10, 0, -1):
    print(i)
```

Print the multiplication table of a given number.

```python
num = int(input("Enter number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```
