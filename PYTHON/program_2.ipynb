Write a function to calculate the total bill amount including 18% GST.

```python
def total_bill(amount):
    gst = amount * 0.18
    total = amount + gst
    return total

bill = float(input("Enter bill amount: "))
print("Total with GST:", total_bill(bill))
```

Create a function that takes a student’s marks list and returns the average and grade.

```python
def result(marks):
    avg = sum(marks) / len(marks)

    if avg >= 90:
        grade = "A"
    elif avg >= 75:
        grade = "B"
    elif avg >= 50:
        grade = "C"
    else:
        grade = "Fail"

    return avg, grade

marks = [80, 70, 90]
avg, grade = result(marks)

print("Average:", avg)
print("Grade:", grade)
```

Write a function to check if a given number is eligible to vote (age ≥ 18).

```python
def vote(age):
    if age >= 18:
        return "Eligible to vote"
    else:
        return "Not eligible"

age = int(input("Enter age: "))
print(vote(age))
```

Create a function that takes a price and discount %, and returns the final price.

```python
def final_price(price, discount):
    discount_amount = price * (discount / 100)
    return price - discount_amount

p = float(input("Enter price: "))
d = float(input("Enter discount %: "))

print("Final price:", final_price(p, d))
```


Write a function to find the smallest number in a list.

```python
def smallest(lst):
    return min(lst)

numbers = [10, 5, 8, 2, 9]
print("Smallest number:", smallest(numbers))
```


Create a function to count how many even numbers are in a list.

```python
def count_even(lst):
    count = 0
    for i in lst:
        if i % 2 == 0:
            count += 1
    return count

numbers = [1, 2, 3, 4, 6, 8]
print("Even numbers count:", count_even(numbers))
```
