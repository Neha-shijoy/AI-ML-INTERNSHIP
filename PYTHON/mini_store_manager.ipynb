DAY-OFF TASK – Mini Store Manager Objective:
 Build a simple system to manage items and prices in a small store.

 1. Create Data 
 • Create a list to store item names 
 • Create a dictionary to store item prices

```python
items = ["pen", "book", "bag", "watch", "shoes"]

```

 2. Add Items 
 • Add at least 5 items with prices items = ["pen", "book", "bag"] 
 prices = {"pen": 10, "book": 50, "bag": 300}

```python
prices = {
    "pen": 10,
    "book": 50,
    "bag": 300,
    "watch": 500,
    "shoes": 800
}
```

  3. Display Items 
  • Print all items with their prices 

```python
for item in items:
    print(item, ":", prices[item])
```

4. Update Price 
  • Change the price of any one item 

```python
prices["book"] = 80
print(prices)
```

5. Remove Item 
  • Remove any one item from list and dictionary

```python
items.remove("pen")
del prices["pen"]

print(items)
print(prices)
```

 6. Find Expensive Item 
  • Print the item with the highest price

```python
expensive_item = max(prices, key=prices.get)
print("Most expensive item:", expensive_item, prices[expensive_item])
```

 7. Calculate Total Price 
 • Find total cost of all items Challenge 
 • Print items with price greater than 100 Bonus
 • Search for an item and display its price Instructions
 • Use list and dictionary methods only 
 • Keep code simple and clean

```python
total = sum(prices.values())
print("Total price:", total)
```

```python
for item, price in prices.items():
    if price > 100:
        print(item, price)
```
