Mini Project: Student Result System with Search 

Develop a console-based Python program that manages student results. The program should allow users to input student names and marks, calculate averages, assign grades based on conditions, display all student results, and provide a search feature to find a specific student by name.

```python
students = []

n = int(input("Enter number of students: "))

# 1. Input student data
for i in range(n):
    print("\nEnter details for Student", i + 1)
    name = input("Name: ")

    m1 = int(input("Subject 1 marks: "))
    m2 = int(input("Subject 2 marks: "))
    m3 = int(input("Subject 3 marks: "))

    total = m1 + m2 + m3
    avg = total / 3

    # 2. Grade calculation
    if avg >= 90:
        grade = "A"
    elif avg >= 75:
        grade = "B"
    elif avg >= 50:
        grade = "C"
    else:
        grade = "Fail"

    # 3. Store in list
    students.append({
        "name": name,
        "total": total,
        "average": avg,
        "grade": grade
    })

# 4. Display all students
print("\n--- Student Results ---")
for s in students:
    print("\nName:", s["name"])
    print("Total:", s["total"])
    print("Average:", s["average"])
    print("Grade:", s["grade"])

# 5. Search student
search = input("\nEnter student name to search: ")

found = False
for s in students:
    if s["name"].lower() == search.lower():
        print("\n--- Student Found ---")
        print("Name:", s["name"])
        print("Total:", s["total"])
        print("Average:", s["average"])
        print("Grade:", s["grade"])
        found = True
        break

if not found:
    print("Student not found")
```
